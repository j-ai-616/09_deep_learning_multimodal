# 컴퓨터 비전의 진화: CNN/Vision Transformer부터 Vision Language Model까지

본 자료는 객체 탐지(Object Detection)와 세그멘테이션(Segmentation)의 기반이 되는 CNN 모델부터 시작하여, 최근 AI 트렌드의 핵심인 Transformer 기반의 ViT, CLIP, BLIP 모델의 원리와 발전 과정을 다룬다. 마지막으로 학습 이후 나아가야 할 최신 로드맵을 제시한다.


## Part 1. CNN 기반의 비전 시스템 (Object Detection & Segmentation)

트랜스포머의 등장에도 불구하고 CNN(Convolutional Neural Network)은 여전히 컴퓨터 비전의 강력한 기둥이다.  
특히 실시간 처리가 중요한 산업 현장에서는 CNN이 주류를 이룬다.

### 1. CNN이 여전히 강력한 이유

* **Inductive Bias (귀납적 편향):** CNN은 이미지가 지역적(Locality) 특성을 가지며, 물체의 위치가 바뀌어도 동일한 물체로 인식해야 한다(Translation Invariance)는 가정을 구조적으로 포함하고 있다. 이로 인해 적은 데이터로도 효율적인 학습이 가능하다.
* **속도와 효율성:** 트랜스포머 대비 연산량이 적고 메모리 효율이 좋아 모바일 및 엣지 디바이스 배포에 유리하다.

### 2. 주요 태스크 및 아키텍처

#### A. 객체 탐지 (Object Detection)

이미지 내의 객체 위치(Bounding Box)를 찾고 종류(Class)를 분류하는 태스크이다.

* **YOLO (You Only Look Once) 시리즈:**
* **특징:** 1-Stage Detector의 대표 주자이다. 속도와 정확도의 균형이 뛰어나 실시간 탐지 분야의 **산업 표준(Industry Standard)**이다.
* **백본 구조:** CSPNet, ELAN 등 효율적인 CNN 구조를 사용하여 특징을 추출한다.
* **최신 동향:** YOLOv8 이후 YOLOv10, YOLO11 등으로 발전하며 실시간 탐지 성능과 배포 효율을 계속 개선하고 있다. 특히 YOLOv10은 NMS-free end-to-end 탐지를 강조하고, YOLO11은 탐지, 세그멘테이션, 포즈 추정, 추적, 분류 등 다양한 비전 태스크를 지원하는 방향으로 정리할 수 있다.

* **하이브리드 모델 (DETR):**
* 객체 탐지를 여러 개의 후보 박스를 정리하는 문제가 아니라, 이미지 안의 객체 집합을 직접 예측하는 set prediction 문제로 바라본다. CNN(ResNet 등) 백본으로 특징을 추출하고 Transformer Encoder-Decoder를 통해 객체 쿼리를 예측하며, 기존 탐지 모델에서 자주 사용하던 NMS(Non-Maximum Suppression) 후처리 의존도를 줄인 것이 특징이다.

#### B. 세그멘테이션 (Segmentation)

픽셀 단위로 객체를 분류하는 정밀한 작업이다.

* **기존 강자 (CNN 기반):** U-Net, Mask R-CNN 등이 있다.
* **최신 트렌드 (Transformer와 Foundation Model 도입):** 세그멘테이션은 픽셀 간의 관계와 전체적인 문맥(Global Context) 파악이 중요하다. Swin Transformer 같은 계층적 Transformer 백본이 널리 활용되었고, 이후에는 SAM, SAM 2처럼 프롬프트 기반으로 이미지와 비디오 객체를 분할하는 foundation model 계열이 중요한 흐름이 되었다.


## Part 2. Vision Transformer와 멀티모달의 시대 (Foundation Models)

자연어 처리(NLP)를 평정한 Transformer 구조가 이미지 처리에 도입되면서, 텍스트와 이미지를 결합한 멀티모달(Multimodal) 모델의 시대가 열렸다. 학습 순서는 **ViT → CLIP → BLIP**을 권장한다.

### 1. ViT (Vision Transformer, 2020)

*CNN의 합성곱 연산을 Self-Attention 메커니즘으로 완전히 대체한 기념비적인 모델이다.*

* **핵심 아이디어:** 이미지를 단어(Token)처럼 취급한다.
1. 이미지를 크기의 **패치(Patch)**로 자른다.
2. 각 패치를 평탄화(Flatten)하여 벡터로 변환한다(Patch Embedding).
3. 패치의 순서 정보를 주기 위해 **Positional Embedding**을 더한다.
4. 이를 Transformer Encoder에 넣어 전체 이미지를 처리한다.

* **의의:** 이미지를 시퀀스 데이터처럼 처리함으로써, 이미지 전체의 문맥(Global Context)을 한 번에 파악할 수 있게 되었다. 최신 대형 모델들의 이미지 인코더(Backbone) 역할을 수행한다.

### 2. CLIP (Contrastive Language-Image Pre-training, 2021)

*이미지와 텍스트를 하나의 벡터 공간에서 이해하게 만든 '연결 고리' 모델이다.*

* **구조:** Two-tower 구조 (이미지 인코더 + 텍스트 인코더).
* **학습 원리 (Contrastive Learning):**
* 인터넷에서 수집한 4억 쌍의 (이미지, 텍스트) 데이터를 사용한다.
* 정답 쌍(Positive pair)의 유사도는 높이고, 오답 쌍(Negative pair)의 유사도는 낮추는 방식으로 학습한다.

* **Zero-shot 능력:** "강아지 사진"이라는 레이블을 학습하지 않아도, 텍스트 인코더가 "강아지"라는 단어의 의미를 알고 있다면 이미지를 분류할 수 있다.
* **한계:** 이미지와 텍스트가 얼마나 유사한지(Matching)는 잘 알지만, 이미지를 보고 텍스트를 생성(Captioning)하지는 못한다.

### 3. BLIP (Bootstrapping Language-Image Pre-training, 2022)

*CLIP의 한계를 넘어, '이해(Understanding)'와 '생성(Generation)'을 통합한 모델이다.*

* **핵심 문제 해결:** 인터넷 데이터의 노이즈(잘못된 이미지 설명)를 해결하고, 생성 능력까지 갖추고자 했다.
* **구조 (MED, Multimodal Encoder-Decoder):**
* **Unimodal Encoder:** 이미지와 텍스트를 각각 이해.
* **Image-grounded Text Encoder:** 이미지 정보를 바탕으로 텍스트를 이해 (VQA 등에 활용).
* **Image-grounded Text Decoder:** 이미지 정보를 바탕으로 텍스트를 생성 (Captioning).

* **CapFilt (Captioning and Filtering):** 데이터 셋을 정제하는 기법이다. 모델이 직접 웹 데이터에 대해 캡션을 생성(Captioner)하고, 생성된 캡션이 이미지와 맞는지 필터링(Filter)하여 고품질 데이터를 확보한다.


## 향후 로드맵: 수업 이후 무엇을 공부해야 하는가?

기초 수업 이후에는 특정 목적(Task)에 맞는 최신 모델(SOTA)을 심도 있게 파고드는 것이 좋다.


### 실무 관점에서의 모델 구분

* **YOLO 계열:** 빠른 객체 탐지에 적합하다. 실시간 웹캠, CCTV, 산업 검사처럼 속도가 중요한 상황에서 많이 사용한다.
* **DETR 계열:** Transformer 기반 객체 탐지 구조를 이해하는 데 중요하다. 객체 탐지를 set prediction 문제로 바라보는 관점을 제공한다.
* **SAM / SAM 2:** 객체의 위치가 아니라 픽셀 단위 마스크를 얻고 싶을 때 사용한다. 점, 박스, 마스크 같은 프롬프트 기반 분할에 강하며, SAM 2는 비디오 객체 분할까지 확장된다.
* **Grounding DINO / YOLO-World:** 텍스트로 찾고 싶은 객체를 지정하는 open-vocabulary detection에 활용한다. 기존 클래스 목록에 없는 객체도 자연어 설명으로 탐지할 수 있다.
* **CLIP / SigLIP:** 이미지와 텍스트를 같은 임베딩 공간에서 비교한다. zero-shot 분류, 이미지 검색, 텍스트 기반 이미지 검색에 활용된다.
* **LLaVA / Qwen-VL 계열:** 이미지를 보고 자연어로 답변하는 VQA, 문서 이미지 이해, 멀티모달 챗봇에 활용된다.

### 1. 챗봇 및 시각적 질의응답 (VQA) 심화

"이 사진 설명해줘", "냉장고 안에 뭐 있어?"와 같은 서비스를 만들고자 한다면 **MLLM (Multimodal LLM)**으로 나아가야 한다.

* **학습 대상:** **LLaVA (Large Language-and-Vision Assistant)**
* **특징:** 이미지 인코더(CLIP/ViT)와 거대 언어 모델(LLaMA 등)을 연결하는 구조(Projection Layer)를 학습한다. LLaVA는 이미지 인코더와 LLM을 연결해 시각 질의응답을 수행하는 대표적인 오픈소스 MLLM 베이스라인으로 볼 수 있다.
* **추가 옵션:** **Qwen-VL/Qwen2.5-VL 계열**, **InternVL 계열** 등도 함께 살펴볼 수 있다. OCR, 문서 이미지 이해, 고해상도 이미지 처리, 시각 질의응답 등에서 자주 활용된다.

### 2. 프롬프트 기반의 차세대 세그멘테이션

기존에 일일이 레이블링하여 학습하던 방식에서, 프롬프트(점, 박스, 텍스트)로 제어하는 방식으로 변화했다.

* **학습 대상:** **SAM (Segment Anything Model)**
* **특징:** 메타(Meta)에서 공개한 모델로, "모든 것을 분할한다"는 목표를 가진다. 대규모 데이터로 사전 학습된 모델을 별도 fine-tuning 없이 사용할 수 있으며, 점, 박스, 마스크 같은 프롬프트를 입력하면 해당 객체의 마스크를 예측한다. 이미지 편집, 데이터 라벨링 보조, 전처리 등에 유용하다. SAM 2는 이러한 프롬프트 기반 분할 방식을 이미지뿐 아니라 비디오까지 확장한 모델로 볼 수 있다.

### 3. Open-Vocabulary 객체 탐지

학습 데이터셋(COCO 등)에 없는 물체도 텍스트 설명만으로 찾아내는 기술이다.

* **학습 대상:** **Grounding DINO**, **YOLO-World**
* **특징:** 텍스트 입력과 이미지 특징을 함께 사용하여, 기존 학습 데이터셋에 고정된 클래스가 아니더라도 자연어로 지정한 객체를 찾을 수 있다. 예를 들어 "검은 모자를 쓴 고양이"처럼 범주명뿐 아니라 속성이 포함된 표현으로도 탐지할 수 있다. 최근에는 Grounding DINO 1.5처럼 성능형 모델과 Edge 배포를 고려한 모델로 확장되는 흐름도 있다.

### 4. 고성능 백본(Backbone) 연구

ViT와 CLIP의 성능을 개선한 최신 인코더들이다.

* **DINOv2:** 레이블 없는 데이터로 자기지도학습을 수행하여, 객체의 깊이(Depth)나 부분적인 특징을 놀랍도록 잘 잡아낸다.
* **SigLIP:** 구글이 제안한 모델로, CLIP의 contrastive loss를 sigmoid loss 방식으로 개선한 계열이다. 큰 배치 크기에 대한 의존을 줄이고 여러 비전-언어 태스크에서 강력한 성능을 보이는 CLIP 계열 대안으로 활용된다. 단, 모든 상황에서 CLIP을 완전히 대체하는 상위 호환이라기보다, 목적과 데이터에 따라 선택할 수 있는 강력한 대안으로 이해하는 것이 좋다.
