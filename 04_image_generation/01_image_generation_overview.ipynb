# 이미지 생성 AI: VAE, GAN, Diffusion 모델

## 1. 생성 모델이란 무엇인가?

- **생성 모델**은 기존 데이터를 학습하여, 새로운 데이터를 만들어내는 AI 모델이다.
- 대표적으로 이미지, 음성, 텍스트 등 다양한 데이터를 생성할 수 있다.
- 이미지 생성 분야에서는 **VAE, GAN, Diffusion** 모델이 많이 활용된다.


## 2. VAE (Variational Autoencoder)

![](https://miro.medium.com/v2/resize:fit:1280/0*dwtvGrRWRAUJuZm4.gif)

**구조와 원리**

- **인코더(Encoder)** 와 **디코더(Decoder)** 로 구성된다.
- 인코더는 입력 이미지를 잠재 공간(latent space)의 확률 분포로 변환한다.
- 디코더는 이 분포에서 샘플링하여 이미지를 재구성한다.

**수식적 설명**

- VAE의 목적 함수:
  $$
  \mathcal{L} = \mathbb{E}_{q_\phi(z|x)}[\log p_\theta(x|z)] - D_{KL}(q_\phi(z|x) \| p(z))
  $$
  - 첫 번째 항: 재구성 손실 (입력과 출력 이미지의 차이)
  - 두 번째 항: 잠재 분포와 정규분포의 차이(KL 발산)를 최소화

- **설명:**  
  - 입력 이미지를 압축하고, 다시 복원하는 과정에서 잠재 공간을 구조화한다.

**특징 및 활용**

- **장점:** 잠재 공간의 구조가 명확, 의미 있는 벡터 조작 가능
- **단점:** 생성 이미지가 GAN에 비해 다소 흐릿함
- **활용 예시:** 손글씨 숫자 생성, 데이터 압축, 이미지 변형 실험 등

## 3. GAN (Generative Adversarial Network)

https://thispersondoesnotexist.com/

https://poloclub.github.io/ganlab/

![](https://miro.medium.com/v2/resize:fit:1040/0*x2RTBmZUtevU0lW7.gif)

**구조와 원리**

- **생성자(Generator)** 와 **판별자(Discriminator)** 두 개의 신경망으로 구성된다.
- 생성자는 랜덤 노이즈에서 이미지를 만들어내고, 판별자는 입력 이미지를 진짜/가짜로 구분한다.
- 두 네트워크가 서로 경쟁(적대적 학습)하면서 더 정교한 이미지를 생성한다.

**수식적 설명**

- GAN의 목적 함수:
  $$
  \min_G \max_D V(D, G) = \mathbb{E}_{x \sim p_{data}(x)}[\log D(x)] + \mathbb{E}_{z \sim p_z(z)}[\log(1 - D(G(z)))]
  $$
  - $G $: 생성자, $D $: 판별자, $ x $: 실제 데이터, $ z $: 노이즈 벡터

- **설명:**  
  - 생성자는 판별자를 속이기 위해 진짜 같은 이미지를 만든다.
  - 판별자는 진짜/가짜를 잘 구분하도록 학습한다.

**특징 및 활용**

- **장점:** 매우 사실적인 이미지 생성 가능
- **단점:** 학습이 불안정, 모드 붕괴(mode collapse) 현상 발생 가능
- **활용 예시:** 가상 인물 얼굴, 예술 이미지, 이미지 변환 등



## 4. Diffusion 모델

![](https://learnopencv.com/wp-content/uploads/2023/01/diffusion-models-unconditional_image_generation.gif)

**구조와 원리**

- **Diffusion 모델**은 이미지를 점진적으로 노이즈로 바꿨다가, 다시 노이즈에서 이미지를 복원하는 과정을 학습한다.
- 수많은 작은 단계를 거쳐 이미지를 생성하므로, 매우 정교하고 다양한 이미지를 만들 수 있다.

**간단한 수식적 설명**

- **정방향 과정:**  
  - 원본 이미지에 점진적으로 노이즈를 추가한다.
- **역방향 과정:**  
  - 노이즈에서 점진적으로 노이즈를 제거하며 이미지를 복원한다.

- **수식 예시:**  
  - $ x_0 $: 원본 이미지, $ x_t $: t단계 노이즈 이미지
  - 정방향: $ x_t = \sqrt{\alpha_t} x_{t-1} + \sqrt{1 - \alpha_t} \epsilon $
  - 역방향: 모델이 $ x_{t-1} $을 예측

**특징 및 활용**

- **장점:** 고해상도, 사실적이고 다양한 이미지 생성 가능, 텍스트-이미지 변환에 강점
- **단점:** 생성 속도가 느릴 수 있음(최근에는 개선됨)
- **활용 예시:** Stable Diffusion, DALL-E, Midjourney 등 최신 이미지 생성 서비스

## 5. 세 모델 비교

| 구분      | VAE                    | GAN                       | Diffusion                                |
| ------- | ---------------------- | ------------------------- | ---------------------------------------- |
| 핵심 아이디어 | 이미지를 압축했다가 다시 복원한다     | 가짜 이미지를 만들고, 진짜인지 판별하게 한다 | 노이즈에서 이미지를 점진적으로 복원한다                    |
| 구조      | 인코더+디코더                | 생성자+판별자                   | 노이즈 추가 과정+노이즈 제거 모델                      |
| 학습 방식   | 재구성 오차와 잠재공간 분포를 함께 학습 | 생성자와 판별자가 경쟁하며 학습         | 이미지에 노이즈를 추가하고, 이를 제거하는 방법을 학습           |
| 이미지 품질  | 비교적 흐릿한 결과가 나올 수 있음    | 선명하고 사실적인 이미지 생성 가능       | 선명하고 다양하며 품질이 높음                         |
| 잠재 공간   | 비교적 잘 정리되어 있음          | 존재하지만 직접 해석하기는 어려움        | 모델에 따라 다르며, Stable Diffusion은 잠재 공간에서 생성 |
| 최근 활용도  | 생성 모델의 기본 개념 이해에 유용    | 고품질 이미지 생성 연구에 활용         | 현재 이미지 생성 분야에서 가장 널리 활용                  |
| 대표 예시   | VAE, Beta-VAE, VQ-VAE  | StyleGAN, BigGAN          | Stable Diffusion, DALL-E                 |
