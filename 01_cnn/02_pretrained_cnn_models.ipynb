**성능지표 Top-n Accuracy, Top-n Error**

1. **Top-1 정확도**:
    - 모델이 예측한 가장 높은 확률의 클래스가 실제 정답인 경우의 비율
    - 예를 들어, Top-1 정확도가 77.1%라면, 모델이 최상위로 예측한 클래스가 77.1% 확률로 정답이라는 의미
2. **Top-5 정확도**:
    - 모델이 예측한 상위 5개의 클래스 중 하나라도 정답인 경우의 비율
    - 예를 들어, Top-5 정확도가 93.3%라면, 모델이 예측한 상위 5개의 클래스 중 하나가 정답일 확률이 93.3%라는 의미
3. **Top-5 오류율**:
    - 상위 5개의 예측 결과 중 어느 하나도 정답을 포함하지 않을 확률. **Top-5 오류율 = 1 - Top-5 정확도**로 계산한다.
    - 예를 들어, Top-5 정확도가 93.3%라면, Top-5 오류율은 6.7%이다. 이는 모델이 상위 5개 예측 결과 중 어느 하나도 정답을 포함하지 않을 확률이 6.7%라는 의미
    

## Transfer Learning을 배우는 이유

CNN은 이미지의 선, 모서리, 질감, 형태와 같은 시각적 특징을 단계적으로 추출한다. 이러한 저수준·중간 수준의 특징은 특정 데이터셋에만 한정되지 않고, 여러 이미지 문제에서 공통적으로 활용될 수 있다.

전이학습(Transfer Learning)은 이 점을 활용한다. 대규모 데이터셋으로 먼저 학습된 모델을 가져온 뒤, 현재 해결하려는 문제에 맞게 일부 계층만 수정하거나 추가 학습한다. 이를 통해 적은 데이터와 짧은 학습 시간으로도 안정적인 성능을 얻을 수 있다.


## ResNet

https://docs.pytorch.org/vision/main/models/resnet.html

ResNet(Residual Network)은 깊은 CNN을 안정적으로 학습하기 위해 residual connection, 즉 잔차 연결을 도입한 모델이다. 일반적인 깊은 신경망은 층이 많아질수록 학습이 어려워질 수 있는데, ResNet은 입력 정보를 뒤쪽 층에 더해주는 방식으로 이러한 문제를 완화한다.

- **종류**:  
  ResNet-18, ResNet-34, ResNet-50, ResNet-101, ResNet-152 등 다양한 깊이의 모델이 있다.

- **사전학습 내용**:  
  torchvision에서 제공하는 ResNet 모델은 ImageNet 데이터셋으로 미리 학습된 가중치를 사용할 수 있다. ImageNet은 약 1,000개 클래스를 가진 대규모 이미지 분류 데이터셋이다.

- **기본 입력과 출력**:  
  ResNet18의 사전학습 모델은 일반적으로 RGB 224x224 이미지를 입력으로 사용하고, ImageNet의 1,000개 클래스에 대한 분류 결과를 출력한다.

- **이번 실습에서의 사용 방식**:  
  ResNet18의 앞부분은 이미지 특징을 추출하는 feature extractor로 사용하고, 마지막 출력층만 SVHN 데이터셋의 10개 숫자 클래스에 맞게 교체한다.


## SVHN: Street View House Numbers

http://ufldl.stanford.edu/housenumbers/

SVHN(Single View House Numbers)은 구글 Street View 이미지에서 수집된 집 번호 숫자 이미지 데이터셋이다. 이번 실습에서는 거리 이미지에서 잘라낸 숫자 이미지를 보고 0부터 9까지의 숫자 중 하나로 분류한다.

- **데이터 형태**: RGB 컬러 이미지
- **원본 이미지 크기**: 32x32
- **클래스 수**: 10개(0~9)
- **문제 유형**: 이미지 분류(Image Classification)

SVHN은 컬러 이미지이므로 ImageNet으로 사전학습된 CNN 모델을 적용하는 예제로 사용하기 적합하다. 다만 ResNet18은 224x224 입력을 기준으로 사전학습되어 있으므로, SVHN의 32x32 이미지를 모델 입력 크기에 맞게 변환해야 한다.

학습 시간을 줄이기 위해 전체 데이터셋 중 일부만 사용한다. 전이학습의 구조를 이해하는 것이 목적이며, 전체 데이터셋을 사용한 장시간 학습은 추가 실험으로 진행할 수 있다.

```python
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

from torchvision import datasets
from torchvision.models import resnet18, ResNet18_Weights
from torch.utils.data import DataLoader, Subset
from torchsummary import summary
from tqdm import tqdm

# 실행 환경에 GPU가 있으면 GPU를 사용하고, 없으면 CPU를 사용한다.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('device:', device)

# 실험 결과를 일정하게 재현하기 위한 난수 고정
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

num_classes = 10
batch_size = 64
num_epochs = 10
learning_rate = 1e-3
```

## 데이터 로드 및 전처리

```python
# ResNet18의 사전 학습 가중치를 가져온다. 모델 가중치 외에도 권장 전처리 정보도 포함 되어 있다.
weight = ResNet18_Weights.DEFAULT
transform = weight.transforms()

# SVHN 데이터셋 로드
train_dataset = datasets.SVHN(
    root='./data', split='train', download=True, transform=transform
)
test_dataset = datasets.SVHN(
    root='./data', split='test', download=True, transform=transform
)
```

```python
train_dataset, test_dataset
```

```python
# 실행 시간을 줄이기 위해 일부 데이터만 사용
train_subset = Subset(train_dataset, range(10000))
test_subset = Subset(test_dataset, range(2000))

# DataLoader
# num_workers=2 : 데이터 로드 작업을 subprocess 2개 병렬 수행
# pin_memory=True : GPU 사용 시 CPU에서 더 빠르게 복사
train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_subset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

print('train size : ', len(train_subset))
print('test size : ', len(test_subset))
```

```python
# SVHN의 원본 데이터
raw_image = train_dataset.data[0]
raw_label = train_dataset.labels[0]

print('원본 이미지 shape : ', raw_image.shape)
print('원본 label : ', raw_label)

# transform 적용 된 데이터
image, label = train_dataset[0]

print('전처리 후 이미지 shape : ', image.shape)
print('전처리 후 label : ', label)
```

## 모델 로드

```python
# ImageNet으로 사전 학습 된 ResNet18을 불러온다.
model = resnet18(weights=weight)

# 기존 RestNet18의 마지막 출력층 확인
print(model.fc)

# 모델 구조 확인
summary(model, (3, 224, 224), device='cpu')
```

```python
# 사전 학습 된 feature extractor 부분은 학습하지 않도록 고정한다.
for param in model.parameters():
    param.requires_grad = False 

# 기존의 fc 계층의 feature 확인
in_features = model.fc.in_features

# 마지막 출력층을 SVHN 클래스 수인 10개 출력으로 교체
model.fc = nn.Linear(in_features, num_classes)

print(model.fc)
summary(model, (3, 244, 244), device='cpu')
```

## 학습
이번 학습에서는 ResNet18 전체를 다시 학습하지 않고, 새로 교체한 마지막 fc 계층만 학습한다. 이를 통해 학습 시간을 줄이면서도 사전학습 모델의 특징 추출 능력을 활용할 수 있다.

```python
criterion = nn.CrossEntropyLoss()

# requires_grad=True인 파라미터만 optimizer에 전달한다.
# 현재는 새로 교체한 fc 계층만 학습 대상이다.
params_to_update = [p for p in model.parameters() if p.requires_grad]
optimizer = optim.Adam(params_to_update, lr=learning_rate)

model = model.to(device)


def train_epoch(model, loader):
    model.train()
    train_loss, train_correct, train_total = 0, 0, 0

    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)

        # 이전 batch의 gradient를 초기화한다.
        optimizer.zero_grad()

        # 순전파: 이미지 입력에 대한 클래스 점수를 계산한다.
        outputs = model(imgs)
        loss = criterion(outputs, labels)

        # 역전파 및 가중치 업데이트
        loss.backward()
        optimizer.step()

        train_loss += loss.item() * imgs.size(0)
        _, predicted = outputs.max(1)
        train_correct += (predicted == labels).sum().item()
        train_total += labels.size(0)

    return train_loss / train_total, train_correct / train_total


def eval_epoch(model, loader):
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0

    # 평가 단계에서는 gradient 계산이 필요하지 않다.
    with torch.no_grad():
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * imgs.size(0)
            _, predicted = outputs.max(1)
            val_correct += (predicted == labels).sum().item()
            val_total += labels.size(0)

    return val_loss / val_total, val_correct / val_total
```

```python
history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': []
}

for epoch in tqdm(range(num_epochs)):
    train_loss, train_acc = train_epoch(model, train_loader)
    val_loss, val_acc = eval_epoch(model, test_loader)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(
        f'Epoch {epoch + 1}/{num_epochs}: '
        f'train_loss={train_loss:.4f} train_acc={train_acc:.4f} '
        f'val_loss={val_loss:.4f} val_acc={val_acc:.4f}'
    )
```

## 학습 결과 확인

```python
# 시각화
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='train_loss')
plt.plot(history['val_loss'], label='val_loss')
plt.legend()
plt.title('Loss')

plt.subplot(1, 2, 2)
plt.plot(history['train_acc'], label='train_acc')
plt.plot(history['val_acc'], label='val_acc')
plt.legend()
plt.title('Accuracy')

plt.tight_layout()
plt.show()
```

```python
# 테스트 데이터 일부에 대해 실제 정답과 모델 예측을 비교
model.eval()

images, labels = next(iter(test_loader))
images, labels = images.to(device), labels.to(device)

with torch.no_grad():
    outputs = model(images)
    _, preds = outputs.max(1)

images = images.cpu()
labels = labels.cpu()
preds = preds.cpu()

# ImageNet 기준 정규화를 되돌려 이미지를 보기 좋게 표시한다.
mean = torch.tensor([0.485, 0.456, 0.406])
std = torch.tensor([0.229, 0.224, 0.225])

plt.figure(figsize=(16, 4))

for i in range(8):
    img = images[i].permute(1, 2, 0)
    img = img * std + mean
    img = torch.clamp(img, 0, 1)

    plt.subplot(1, 8, i + 1)
    plt.imshow(img)
    plt.title(f'T:{labels[i].item()} / P:{preds[i].item()}')
    plt.axis('off')

plt.show()
```
